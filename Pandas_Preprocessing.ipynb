### **Data Cleaning**

***>>> country_wise_latest:***  *NO preprocessing needed.*

***>>> day_wise:***  *Change date to datetime.*

***>>> full_grouped:***  *Change date to datetime.*

***>>> usa_county_wise:***  *Change date to datetime && Missing values handling.*

***>>> worldometer_data:***  *Missing values handling.*

#### **Changing date to datetime**

```python
# covid_19_clean_complete
c19cc.Date = pd.to_datetime(c19cc.Date)

# day_wise
dw.Date = pd.to_datetime(dw.Date)

# full_grouped
fg.Date = pd.to_datetime(fg.Date)

# usa_county_wise
usacw.Date = pd.to_datetime(usacw.Date)
```

#### **Missing values handling**

```python
# covid_19_clean_complete

# Drop Province/State Column as it has 70% missing values
c19cc.drop(['Province/State'], axis=1, inplace=True)
```

```python
# usa_county_wise

# Drop rows with missing values
usacw.dropna(subset=['FIPS', 'Admin2'], inplace=True) 
```

```python
# worldometer_data

# Fill missing values with median
wmd['Serious,Critical'] = wmd['Serious,Critical'].fillna(wmd['Serious,Critical'].median)

# Replace missing values in these columns with 0
wmd.fillna({'NewCases':0, 'NewDeaths':0, 'NewRecovered':0}, inplace=True)

# Fill missing values with 'Unkown
wmd['WHO Region'] = wmd['WHO Region'].fillna('Unknown')

# Drop rows with missing values
wmd.dropna(inplace=True)
```

```python
print(f"{cwl.isnull().sum()} \n")

print(f"{c19cc.isnull().sum()} \n")

print(f"{dw.isnull().sum()} \n")

print(f"{fg.isnull().sum()} \n")

print(f"{usacw.isnull().sum()} \n")

print(f"{wmd.isnull().sum()} \n")
```

### **Grouping and Sorting**

```python
# worldometer_data

# reordering the columns 
cols = list(wmd.columns)
cols.remove('WHO Region')
cols.insert(1, 'WHO Region')
wmd = wmd[cols]

# Group by 'WHO Region' and sort within each group by 'Country/Region'
cols = list(wmd.columns)
cols.remove('WHO Region')
wmd = (wmd.groupby('WHO Region', group_keys=True).apply(lambda x: x[cols].sort_values(by='Country/Region')))

print(f"{wmd} \n")
wmd.head()
```
